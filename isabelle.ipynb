Example: Changing a json to the specified format

````python
import json

json_path = 'input/INTEL052V3S.json'
formatted_path = 'input/INTEL052V3S_Formatted.json'
````

````python
import json

with open(json_path, 'r') as f:
    d = json.load(f)

output = []
for item in d['results']['audio_segments']:
    output.append({
        'start': item['start_time'],
        'end': item['end_time'],
        'text': item['transcript'],
    })

with open(formatted_path, 'w') as f:
    json.dump(output, f)
````

````python
from pydub import AudioSegment

# Source - https://stackoverflow.com/q/76618750
# Posted by Hesam Hakemi, modified by community. See post 'Timeline' for change history
# Retrieved 2026-05-08, License - CC BY-SA 4.0

m4a_file = 'input/INTEL052V3S.m4a'
wav_filename = 'input/INTEL052V3S_Formatted.wav'

sound = AudioSegment.from_file(m4a_file, format='m4a')
sound.export(wav_filename, format='wav')
````

Example: Alignment
- Note: I did some debugging here so if it doesn't work let me know.

````python
import AlignPipeline as AP
from lingua import Language
import json

# Set variables & specify tier with utterances
audio_path = 'input/INTEL052V3S_Formatted.wav'
json_path = 'input/INTEL052V3S_Formatted.json'
output_dir = 'output/'

# Load transcript & define what languages are in dataset
with open(json_path, 'r') as f:
    transcript = json.load(f)
languages = [Language.ENGLISH]

# Call script
segments = AP.script(audio_path=audio_path, transcript=transcript, temp_dir=output_dir, languages=languages, download_models=False)
print(segments)
````
