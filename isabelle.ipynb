# 4) For loop

````python
# INTEL080V1_parent_transcript_edit.json
# INTEL080V1S.wav

import AlignPipeline as AP
import EditPipeline as EP
from lingua import Language
import json
import os

directory = 'input/'
output_dir = 'output/'
languages = [Language.ENGLISH]
file2aligned = {}

for file in os.listdir(directory):

    file = directory + file

    # Check if it's a json
    if ".json" in file:
        json_path = file
        audio_path = file.split('_')[0] + "S.m4a"
        wav_path = EP.m4a_to_wav(audio_path)
        print("[FOR-LOOP] Currently:", json_path, wav_path)

        if os.path.isfile(json_path) and os.path.isfile(wav_path):
    
            # Load transcript & define what languages are in dataset
            with open(json_path, 'r') as f:
                transcript = json.load(f)
            
            # Call script
            segments = AP.script(audio_path=audio_path, transcript=transcript, temp_dir=output_dir, languages=languages, download_models=False)
            file2aligned[file] = segments

        else:
            raise Exception("One of those paths doesn't exist")
````
